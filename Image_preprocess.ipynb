```python
import nibabel as nib

# Load the .mgz file
file_path = 'C:/Users/nahid/jupyter/ADNI-onehalfT-FastSurfer-QuickSeg/AD/I31143nii/mri/orig/a.mgz'
img = nib.load(file_path)

# Access the image data as a numpy array
data = img.get_fdata()
print("Data shape:", data.shape)
```

```python
import matplotlib.pyplot as plt

# Plot a slice along each axis
plt.figure(figsize=(12, 4))

# Sagittal slice (left-right view)
plt.subplot(1, 3, 1)
plt.imshow(data[100, :, :], cmap='gray')
plt.title('Sagittal Slice')

# Coronal slice (front-back view)
plt.subplot(1, 3, 2)
plt.imshow(data[:, 100, :], cmap='gray')
plt.title('Coronal Slice')

# Axial slice (top-bottom view)
plt.subplot(1, 3, 3)
plt.imshow(data[:, :, 100], cmap='gray')
plt.title('Axial Slice')

plt.show()
```
